Me devuelve 1803 filas. Definitivamente hay que arreglar los genres. Una solución pausible sería hacer un split y dejar el primer elemento como genre y los siguientes como subgenre y ver como queda. El problema que veo es que no todas las separaciones están hechas igual, unas van con ',' y otras con '-'. Vamos a unificar separaciones y hacer que todo se separe con '-'.

Vamos a usar apply.

```python
def sustituir(genero):
    return genero.replace("," , "-")

df_books["author_genres"]=df_books["author_genres"].apply(sustituir)
df_books.head()
```

Una vez que tenemos los géneros separados solo por '-', vamos a separar la primera palabra del resto y crear dos columnas de subgénero dada la variedad de información.
