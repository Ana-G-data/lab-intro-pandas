Hay varias columnas que no nos interesa mantener, por lo que vamos a crear un nuevo DataFrame con las columnas con las que vamos a trabajar.

Vamos a hacerlo con el método drop. Podemos utilizar el metodo inplace=True o bien asignarle otra variable, que es lo que voya  hacer en este caso.

```python
eliminar=["Author_id","Author_page_url","Book_fullurl", "Book_id"]
df_books2=df_books.drop(eliminar, axis = 1)
df_books2.head(2)
```

Author gender utiliza dos tipos de separadores: "," y "-". Vamos a unificarlo todo con "-"

```python
def sustituir(genero):
    return genero.replace("," , "-")

df_books2["Author_genres"]=df_books2["Author_genres"].apply(sustituir)
df_books2.head(2)
```

De las columnas Author_name, Birthplace y Book_title vamos a eliminar las "\n"

```python
def eliminar(n):
    return n.replace("\n","")

df_books2["Author_name"]=df_books2["Author_name"].apply(eliminar)
df_books2["Book_title"]=df_books2["Book_title"].apply(eliminar)
df_books2["Birthplace"]=df_books2["Birthplace"].apply(eliminar)
df_books2.head(2)
```

La columna de fecha parece que tiene diferentes maneras de mostrarla. Vamos a verlo:

```python
df_books2["Publish_date"]
```

Tenemos las fechas como:
- años 
- meses y años  
- mes, día y año

Quiero quedarme solo con los años. Vamos a utilizar Regex para hacerlo de una manera mucho más sencilla.

```python
df_books2["Publish_date"] = df_books2["Publish_date"].str.extract(r'(\d{4})')
df_books2.head()
```

Genre_1 y Genre_2 tenemos 124 y 157 valores únicos, vamos a unificarlos un poco a ver cuanto podemos bajar ese número.

```python
def sustituir(genre):
    if "Art" in genre:
        return "Art" 
    elif "War" in genre:
        return "War"
    elif "Fiction" in genre:
        return "Fiction"
    elif "Comic" in genre:
        return "Comic"
    elif "Glbt" in genre:
        return "Lgbt"
    elif "Romance" in genre:
        return "Romance"
    elif "Fantasy" in genre:
        return "Fantasy"
    elif "Finish" in genre:
        return "Unfinished"
    elif "Thriller" in genre:
        return "Thriller"
    elif "Novella" in genre:
        return "Novels"
    elif "Literature" in genre:
        return "Literature"
    elif "Sports" in genre:
        return "Sports"   
    elif "Christian" in genre:
        return "Christianity"

    else:
        return genre
   


df_books2["Genre_1"]=df_books2["Genre_1"].apply(sustituir)
df_books2["Genre_2"]=df_books2["Genre_2"].apply(sustituir)
df_books2.describe(include = "object").T


```

Hay veces que los títulos cortos y largos se ponen de moda en diferentes años a raíz de un Bestseller, vamos a ver si hay algun patrón de ello, vamos a usar los años desde el 2000.

```python
def palabras (titulo):
    if len(titulo.split())<=2:
        return "Corto"
    else: 
        return "Largo"

df_books2["Len_título"]=df_books2["Book_title"].apply(palabras)

df_books2["Len_años"]=""

df_len_años=df_books2.groupby(['Publish_date','Len_título'])["Len_años"].count().reset_index()

df_len_años['Publish_date'] = df_len_años['Publish_date'].astype("int64")
df_len_años.loc[(df_len_años['Publish_date'] >= 2000)]
```
